```python
import pandas as pd
import json
```

```python
with open('../veri/ankarabbSoruCevapVeriSetiJson-v02.json', encoding='utf-8') as f:
    veri = json.load(f)
```

```python
veri
```

```python
first_list = list()
first_dict = dict()
for veri_ in veri:
    # print(veri_)
    for index in veri[veri_]:
        print(index)
        new_dict = dict()
        new_dict["tag"] = index["etiket"]
        new_dict["patterns"] = [index["soru"]]
        new_dict["responses"] = [index["cevap"]]
        first_list.append(new_dict)
    first_dict["Tüm Sorular"] = first_list
```

```python
first_dict["Tüm Sorular"]
```

```python
with open('../veri/ankarabbSoruCevapVeriSetiJson-v03.json', 'w', encoding='utf-8') as json_dosya:
    json.dump(first_dict, json_dosya, ensure_ascii=False)
```
