```python
import json
import numpy as np
import os
```

```python
m = 0
writer = False

data = {}
allQuestions = []
index = {}
indexTag = str()
indexPatterns = []
indexResponses = []

for i,j in enumerate(df.values[:-1]):
    if str(j[0]) != "nan" and str(j[2]) != "nan":
        indexTag = j[0]
        indexResponses.append(j[2])
        writer = False
    if str(df.values[i+1][0]) != "nan" and str(df.values[i+1][2]) != "nan":
        writer = True
        indexPatterns.append(j[1])
    else:
        writer= False
        indexPatterns.append(j[1])
    
    if writer:
        index["tag"] = indexTag
        index["patterns"] = indexPatterns
        index["responses"] = indexResponses
        allQuestions.append(index)
        indexPatterns = []
        index = {}
        indexTag = str()
        indexResponses = []
data["Tüm Sorular"] = allQuestions

data
```

```python
path = r'C:\Users\alierenekinci\Desktop\Project\ThreadTeknofest\veri'

def kaydet(veri):
    liste = os.listdir(path)
    final_path = path + "\\" + liste[-1][:-7] + str(int(liste[-1][31:-5])+1) + ".json"
    with open(final_path, 'w', encoding='utf-8') as json_dosya:
        json.dump(veri, json_dosya, ensure_ascii=False)

    print("Dosya adı:", final_path)
```
